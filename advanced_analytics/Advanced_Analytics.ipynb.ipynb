```python
nav['daily_return'] = nav.groupby('amfi_code')['nav'].pct_change()
```

```python
nav.head(10)
```

```python
var = nav.groupby('amfi_code')['daily_return'].apply(
    lambda x: x.dropna().quantile(0.05)
)
```

```python
cvar = nav.groupby('amfi_code')['daily_return'].apply(
    lambda x: x[x <= x.quantile(0.05)].mean()
)
```

```python
var_cvar_report = pd.DataFrame({
    'VaR_95': var,
    'CVaR_95': cvar
}).reset_index()
```

```python
var_cvar_report = var_cvar_report.merge(
    fund_master[['amfi_code', 'scheme_name']],
    on='amfi_code',
    how='left'
)
```

```python
nav['rolling_sharpe'] = (
    nav.groupby('amfi_code')['daily_return']
       .transform(lambda x: (x.rolling(90).mean() / x.rolling(90).std()) * np.sqrt(252))
)
```

```python
top5 = fund_master['amfi_code'].head(5)
```

```python
plt.figure(figsize=(12,6))

for fund in top5:
    data = nav[nav['amfi_code'] == fund]
    name = fund_master.loc[
        fund_master['amfi_code'] == fund,
        'scheme_name'
    ].values[0]

    plt.plot(data['date'], data['rolling_sharpe'], label=name)

plt.title("90-Day Rolling Sharpe Ratio")
plt.xlabel("Date")
plt.ylabel("Sharpe Ratio")
plt.legend()
plt.grid(True)

plt.savefig("rolling_sharpe_chart.png")
plt.show()
```

```python
first_transaction = transactions.groupby('investor_id')['transaction_date'].min().reset_index()
```

```python
first_transaction.head()
```

```python
first_transaction['cohort_year'] = first_transaction['transaction_date'].dt.year
```

```python
first_transaction.head()
```

```python
transactions = transactions.merge(
    first_transaction[['investor_id', 'cohort_year']],
    on='investor_id',
    how='left'
)
```

```python
average_investment = transactions.groupby('cohort_year')['amount_inr'].mean().reset_index()

average_investment.rename(columns={'amount_inr':'average_investment'}, inplace=True)

print(average_investment)
```

```python
total_investment = transactions.groupby('cohort_year')['amount_inr'].sum().reset_index()

total_investment.rename(columns={'amount_inr':'total_investment'}, inplace=True)

print(total_investment)
```

```python
transactions['transaction_type'].value_counts()
```

```python
sip_transactions = transactions[transactions['transaction_type'] == 'SIP']

average_sip = sip_transactions.groupby('cohort_year')['amount_inr'].mean().reset_index()

average_sip.rename(columns={'amount_inr':'average_sip_amount'}, inplace=True)

average_sip
```

```python
investments = transactions[
    transactions['transaction_type'].isin(['SIP', 'Lumpsum'])
]

total_investment = investments.groupby('cohort_year')['amount_inr'].sum().reset_index()

total_investment.rename(columns={'amount_inr':'total_investment'}, inplace=True)

total_investment
```

```python
transactions = transactions.merge(
    fund_master[['amfi_code', 'scheme_name']],
    on='amfi_code',
    how='left'
)
```

```python
top_fund = (
    investments.merge(
        fund_master[['amfi_code', 'scheme_name']],
        on='amfi_code'
    )
    .groupby(['cohort_year', 'scheme_name'])
    .size()
    .reset_index(name='count')
)

top_fund = top_fund.loc[
    top_fund.groupby('cohort_year')['count'].idxmax()
]

top_fund
```

```python
cohort_report = average_sip.merge(
    total_investment,
    on='cohort_year'
)

cohort_report = cohort_report.merge(
    top_fund[['cohort_year', 'scheme_name']],
    on='cohort_year'
)

cohort_report.rename(columns={
    'scheme_name': 'top_fund'
}, inplace=True)

cohort_report
```

```python
sip_transactions = transactions[transactions['transaction_type'] == 'SIP'].copy()
```

```python
sip_transactions = sip_transactions.sort_values(
    ['investor_id', 'transaction_date']
)
```

```python
sip_transactions['gap_days'] = sip_transactions.groupby('investor_id')[
    'transaction_date'
].diff().dt.days
```

```python
sip_count = sip_transactions.groupby('investor_id').size()
```

```python
eligible_investors = sip_count[sip_count >= 6].index
```

```python
sip_transactions = sip_transactions[
    sip_transactions['investor_id'].isin(eligible_investors)
]
```

```python
average_gap = (
    sip_transactions.groupby('investor_id')['gap_days']
    .mean()
    .reset_index()
)
```

```python
average_gap.rename(
    columns={'gap_days': 'average_gap_days'},
    inplace=True
)
```

```python
average_gap['status'] = average_gap['average_gap_days'].apply(
    lambda x: 'At Risk' if x > 35 else 'Active'
)
```
