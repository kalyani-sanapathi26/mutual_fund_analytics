```python
# Cell 3: Fund Scorecard - Composite Ranking (no function def to avoid indent issues)
df['score_return_3yr'] = df['return_3yr_pct'].rank(pct=True) * 100
df['score_sharpe'] = df['sharpe_ratio'].rank(pct=True) * 100
df['score_alpha'] = df['alpha'].rank(pct=True) * 100
df['score_expense'] = (1 - df['expense_ratio_pct'].rank(pct=True)) * 100
df['score_drawdown'] = (1 - df['max_drawdown_pct'].rank(pct=True)) * 100
df['composite_score'] = (0.30*df['score_return_3yr'] + 0.25*df['score_sharpe'] + 0.20*df['score_alpha'] + 0.15*df['score_expense'] + 0.10*df['score_drawdown']).round(2)
df['rank'] = df['composite_score'].rank(ascending=False).astype(int)
df_sorted = df.sort_values('rank')
scorecard_cols = ['rank','scheme_name','fund_house','category','plan','return_3yr_pct','sharpe_ratio','alpha','expense_ratio_pct','max_drawdown_pct','morningstar_rating','composite_score']
fund_scorecard = df_sorted[scorecard_cols].reset_index(drop=True)
print('=== TOP 10 FUNDS BY COMPOSITE SCORE ===')
print(fund_scorecard.head(10).to_string(index=False))
fund_scorecard.to_csv('fund_scorecard.csv', index=False)
print('fund_scorecard.csv saved! Total funds:', len(fund_scorecard))
```

```python
# Cell 4: Alpha & Beta Summary CSV
alpha_beta = df[['amfi_code','scheme_name','fund_house','category','plan','alpha','beta','return_3yr_pct','benchmark_3yr_pct']].copy()
alpha_beta['outperformance'] = (alpha_beta['return_3yr_pct'] - alpha_beta['benchmark_3yr_pct']).round(2)
alpha_beta = alpha_beta.sort_values('alpha', ascending=False).reset_index(drop=True)
print('=== ALPHA & BETA SUMMARY (Top 10 by Alpha) ===')
print(alpha_beta[['scheme_name','alpha','beta','outperformance']].head(10).to_string(index=False))
alpha_beta.to_csv('alpha_beta.csv', index=False)
print('alpha_beta.csv saved! Rows:', len(alpha_beta))
```

```python
# Cell 5: Benchmark Comparison Chart - Top 5 Funds vs Nifty 50 & Nifty 100 (3yr returns)
top5 = fund_scorecard.head(5).copy()
nifty50_3yr = 11.8
nifty100_3yr = 12.3
scheme_names = [s.replace(' - Regular Plan','').replace(' - Growth','').replace(' - Regular','')[:30] for s in top5['scheme_name'].tolist()]
return_vals = top5['return_3yr_pct'].tolist()
all_names = scheme_names + ['Nifty 50 Index', 'Nifty 100 Index']
all_returns = return_vals + [nifty50_3yr, nifty100_3yr]
colors = ['#2196F3','#4CAF50','#FF9800','#9C27B0','#F44336','#607D8B','#795548']
fig, ax = plt.subplots(figsize=(14,7))
bars = ax.bar(range(len(all_names)), all_returns, color=colors, alpha=0.85, edgecolor='white', linewidth=1.2)
ax.axhline(y=nifty50_3yr, color='#607D8B', linestyle='--', linewidth=1.5, alpha=0.7, label=f'Nifty 50 ({nifty50_3yr}%)')
ax.axhline(y=nifty100_3yr, color='#795548', linestyle=':', linewidth=1.5, alpha=0.7, label=f'Nifty 100 ({nifty100_3yr}%)')
for i, (bar, val) in enumerate(zip(bars, all_returns)):
 ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.2, f'{val:.1f}%', ha='center', va='bottom', fontsize=10, fontweight='bold')
ax.set_xticks(range(len(all_names)))
ax.set_xticklabels(all_names, rotation=35, ha='right', fontsize=9)
ax.set_ylabel('3-Year Returns (%)', fontsize=12)
ax.set_title('Top 5 Funds vs Benchmark (Nifty 50 & Nifty 100) - 3 Year Returns', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.set_ylim(0, max(all_returns)*1.15)
ax.grid(axis='y', alpha=0.3)
fig.tight_layout()
fig.savefig('benchmark_comparison_chart.png', dpi=150, bbox_inches='tight')
plt.show()
print('benchmark_comparison_chart.png saved!')
```
